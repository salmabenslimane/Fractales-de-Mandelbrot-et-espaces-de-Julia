#**A. Fractale de Mandelbrot:**

##**A.1.a. Description du phénomène étudié :**

La fractale de Mandelbrot est une fractale mathématique qui affiche une structure complexe et infiniment détaillée lorsqu'elle est visualisée. Elle doit son nom au mathématicien français Benoît Mandelbrot, qui l'a étudiée et popularisée dans les années 1970.

La fractale de Mandelbrot est principalement étudiée dans le domaine des mathématiques et de la visualisation graphique. Cependant, il existe des domaines scientifiques où les fractales, y compris la fractale de Mandelbrot, peuvent être utilisées pour modéliser des phénomènes naturels ou pour analyser des données complexes. L'un de ces domaines est la théorie du chaos.

La théorie du chaos étudie les systèmes dynamiques complexes qui présentent une sensibilité extrême aux conditions initiales. Ces systèmes peuvent exhiber des comportements chaotiques, c'est-à-dire des trajectoires imprévisibles et hautement sensibles aux petites variations des conditions initiales.

La fractale de Mandelbrot est souvent utilisée pour illustrer et étudier les propriétés du chaos. En particulier, elle montre comment une simple formule itérative peut générer une structure complexe et infiniment détaillée. La frontière de la fractale de Mandelbrot elle-même représente une région où le comportement du système est complexe et chaotique.

La fractale de Mandelbrot peut être utilisée comme un outil visuel pour explorer et comprendre les concepts clés de la théorie du chaos, tels que la sensibilité aux conditions initiales, les attracteurs étranges et les bifurcations. Elle permet d'illustrer comment de petites variations dans les conditions initiales peuvent entraîner des résultats complètement différents et comment des motifs répétitifs et complexes peuvent émerger à partir de règles simples.



##**A.1.b. Mise en équation :**

La fractale de Mandelbrot est générée à partir d'une formule mathématique simple, utilisant les nombres complexes. On définit ainsi une suite itérative qui est donnée par la relation récursive suivante :


\begin{cases}
z_0=0\\
z_{n+1}=z_n^2+c
\end{cases}



Dans cette équation, $z[n]$ est un nombre complexe représentant l'état à l'instant $n$, $z[n+1]$ est l'état suivant à l'instant $n+1$, et $c$ est un nombre complexe constant correspondant aux coordonnées du point à l'étude dans le plan complexe.

La fractale de Mandelbrot est créée en examinant le comportement de cette suite pour chaque point du plan complexe. Pour chaque point $c$, on commence avec $z₀ = 0 $ et on itère la relation récursive. Si la suite diverge et tend vers l'infini, le point $c$ est considéré comme n'appartenant pas à la fractale de Mandelbrot. En revanche, si la suite reste bornée pour toutes les itérations, le point c est considéré comme appartenant à la fractale.

###A.2.1 Étude de l'existence et de l'unicité de la solution:

La fractale de Mandelbrot est créée en examinant le comportement de cette suite pour chaque point du plan complexe. Pour chaque point $c$, on commence avec $z₀ = 0$ et on itère la relation récursive. Si la suite diverge et tend vers l'infini, le point $c$ est considéré comme n'appartenant pas à la fractale de Mandelbrot. En revanche, si la suite reste bornée pour toutes les itérations, le point $c$ est considéré comme appartenant à la fractale.

Pour étudier l'existence et l'unicité des solutions, nous allons analyser la convergence ou la divergence des trajectoires générées par l'équation.

***Existence de la solution :***

L'existence de la solution pour le système dynamique de la Fractale de Mandelbrot est déterminée en évaluant si la séquence $z[n]$ converge ou diverge.

*Théorème de l'ensemble de Mandelbrot :*

Pour chaque point $c$ dans le plan complexe, la séquence $z[n]$ est bornée lorsque $n$ tend vers l'infini (c'est-à-dire qu'elle ne tend pas vers l'infini) ou qu'elle diverge. Ainsi, une solution existe pour chaque point du plan complexe.

On veut déterminer si la suite $z[n]$ reste bornée pour un point $c$ donné. Pour cela, appliquons le théorème de la compacité. Avant supposons que la suite $z[n]$ est bornée. Cela signifie qu'il existe une constante $M$ telle que pour tout $n$, $|z[n]| ≤ M$.

Ensuite, utilisons le théorème de la compacité pour affirmer qu'il existe une sous-suite de $z[n]$, notée $z[n_k]$, qui converge vers une limite, notée $L$.

Si la limite $L$ est finie, cela implique que la suite $z[n]$ converge vers cette limite $L$. Cela signifie que le point $c$ appartient à l'ensemble de Mandelbrot.

En revanche, si la limite $L$ est infinie, cela implique que la suite $z[n]$ diverge vers l'infini. Dans ce cas, le point $c$ n'appartient pas à l'ensemble de Mandelbrot.

En résumé, le théorème de la compacité est utilisé pour montrer l'existence des solutions dans l'ensemble de Mandelbrot en montrant que toute suite $z[n]$ bornée contient une sous-suite convergente.

Il est important de noter que le théorème de la compacité ne garantit pas l'unicité de la solution.


**Unicité de la solution :**

L'unicité de la solution fait référence à la possibilité d'avoir plusieurs trajectoires distinctes convergentes ou divergentes pour un même point du plan complexe.

*Théorème de l'unicité de la solution :*

Dans le cas de la Fractale de Mandelbrot, chaque point du plan complexe $c$ génère une trajectoire unique qui est soit convergente, soit divergente. Il n'y a pas de cas où deux trajectoires distinctes à partir d'un même point convergent vers des valeurs différentes.

**Preuve par contradiction :**

Supposons qu'il existe deux trajectoires distinctes à partir d'un même point $c$, qui convergent vers des valeurs différentes, $z1$ et $z2$, respectivement. Cela signifierait que $z1[n]$ et $z2[n]$ restent à une distance constante l'une de l'autre à mesure que n tend vers l'infini. Cependant, cela contredirait le fait que les trajectoires de la Fractale de Mandelbrot soient déterminées uniquement par les coordonnées du point $c$.

En conclusion, pour le système dynamique de la Fractale de Mandelbrot, chaque point du plan complexe a une solution unique, soit une trajectoire convergente, soit une trajectoire divergente. L'existence et l'unicité des solutions sont déterminées par l'analyse de la convergence ou de la divergence des trajectoires générées par l'équation.

###A.2.2. Étude de la dépendance continue par rapport à la condition initiale et à la fonction f:

**A.2.2.1. Dépendance continue par rapport à la condition initiale :**

Pour le système dynamique de la Fractale de Mandelbrot, la dépendance continue par rapport à la condition initiale signifie que de petites variations dans la condition initiale conduisent à des trajectoires proches ou similaires.

**Théorème de continuité pour les systèmes dynamiques :**

Dans le cas de la Fractale de Mandelbrot, de légères variations dans les coordonnées du point initial C entraînent des trajectoires qui restent proches l'une de l'autre. Cela signifie que si deux points $c1$ et $c2$ sont proches dans le plan complexe, les trajectoires correspondantes $z1[n]$ et $z2[n]$ seront également proches.

**Preuve :**

Soient $c1$ et $c2$ deux points proches dans le plan complexe tels que $|c1 - c2|$ est très petit. Supposons que $z1[n]$ et $z2[n]$ soient les trajectoires correspondantes à $c1$ et $c2$, respectivement, et montrons ceci par récurrence:

Pour $n=0$ , on a déjà $|c1 - c2|$ très petit, donc c'est vérifié pour $n=0$

Soit $1≤n$  Supposons  maintenant que  $|z1[n-1] - z2[n-1]|$ est petit, et montrons ceci pour $n$.

En utilisant l'équation récurrente du système dynamique de la Fractale de Mandelbrot, nous pouvons montrer que $|z1[n] - z2[n]|$ reste petit pour  $n$ :

$|z1[n] - z2[n]| = |(z1[n-1]^2 + c1) - (z2[n-1]^2 + c2)|$

$                = |z1[n-1]^2 - z2[n-1]^2 + (c1 - c2)|$
$= |(z1[n-1] - z2[n-1]) * (z1[n-1] + z2[n-1]) + (c1 - c2)|$
$≤ |z1[n-1] - z2[n-1]| * |z1[n-1] + z2[n-1]| + |c1 - c2|$


Puisque $|z1[n-1] - z2[n-1]|$ est petit pour un $n-1$ donné, $|z1[n-1] + z2[n-1]|$ est également borné. De plus, $|c1 - c2|$ est très petit.

Par conséquent, $|z1[n] - z2[n]|$ reste petit pour tout $n$, ce qui confirme la dépendance continue par rapport à la condition initiale.

**A.2.2.2. Dépendance continue par rapport à la fonction f :**

La dépendance continue par rapport à la fonction f dans le système dynamique de la Fractale de Mandelbrot se réfère à la manière dont de légères variations dans la fonction $f(z) = z^2 + c$ affectent les trajectoires générées.

**Théorème de continuité par rapport à la fonction f :**

De petites variations dans la fonction $f(z) = z^2 + c$ n'entraînent que de petites variations dans les trajectoires générées par le système dynamique de la Fractale de Mandelbrot.

**Preuve:**

Soit $f1(z) = z^2 + c1$ et $f2(z) = z^2 + c2$ deux fonctions légèrement différentes, où $|c1 - c2|$ est très petit. Supposons que $z1[n]$ et $z2[n]$ soient les trajectoires correspondantes à $f1(z)$ et $f2(z)$, respectivement.

En utilisant l'équation récurrente du système dynamique, nous pouvons montrer que $|z1[n] - z2[n]|$ reste petit pour tout $n$ :

$|z1[n] - z2[n]| = |f1(z1[n-1]) - f2(z2[n-1])|$
$≤ |z1[n-1] - z2[n-1]| + |c1 - c2|$

Puisque $|z1[n-1] - z2[n-1]|$ est petit pour un $n-1$ donné et $|c1 - c2|$ est très petit, $|z1[n] - z2[n]|$ reste petit pour tout $n$.

Ainsi, les trajectoires générées par des fonctions légèrement différentes restent proches les unes des autres, confirmant la dépendance continue par rapport à la fonction $f$.

En conclusion, le système dynamique de la Fractale de Mandelbrot présente une dépendance continue par rapport à la condition initiale et à la fonction f. De petites variations dans la condition initiale ou dans la fonction f entraînent des trajectoires proches ou similaires, confirmant la régularité et la stabilité du système dynamique.

###A.2.3. Étude de la stabilité des points fixes:

Un point fixe dans le système dynamique de la Fractale de Mandelbrot est un point où la trajectoire reste constante et ne change pas au fil du temps. Pour étudier la stabilité des points fixes, nous analysons comment de petites perturbations autour de ces points se propagent dans le système.

Considérons un point fixe $Z*$. Mathématiquement, cela signifie que $Z*[n+1] = Z*[n]^2 + C$ pour tout $n$, où $Z*[n]$ est la trajectoire du point fixe à l'instant $n$ et $C$ est le paramètre constant du point étudié.

Pour étudier la stabilité du point fixe, nous évaluons la dérivée du système dynamique au voisinage du point fixe et analysons sa valeur.

Soit $f(z) = z^2 + C$ la fonction du système dynamique de la Fractale de Mandelbrot. Calculons la dérivée de $f(z)$ par rapport à $z$ :

$f'(z) = 2z$

À un point fixe $Z*$, $f'(Z*) = 2Z*$.

Maintenant, examinons les différents cas possibles :

Si $|f'(Z*)| < 1$ :
Cela signifie que la dérivée au point fixe est strictement inférieure à 1 en valeur absolue. Dans ce cas, les perturbations autour du point fixe diminueront au fil du temps, et le point fixe est considéré comme stable.

Si $|f'(Z*)| > 1$ :
Si la dérivée au point fixe est strictement supérieure à 1 en valeur absolue, les perturbations autour du point fixe augmenteront au fil du temps, et le point fixe est considéré comme instable.

Si $|f'(Z*)| = 1$ :
Si la dérivée au point fixe est égale à 1 en valeur absolue, une conclusion définitive sur la stabilité ne peut pas être tirée uniquement sur la base de la dérivée. Dans ce cas, des analyses supplémentaires sont nécessaires pour déterminer la stabilité.

Il convient de noter que dans le système dynamique de la Fractale de Mandelbrot, les points fixes peuvent être attractifs (stables), répulsifs (instables) ou donner lieu à des comportements plus complexes tels que des cycles limites ou des comportements chaotiques.

###A.2.4. Étude des portraits de phases

**Portrait de phase pour le problème linéarisé :**

Dans le problème linéarisé, nous considérons l'équation récurrente simplifiée $Z[n+1] = Z[n] + C$, où $C$ est une constante.
Pour tracer le portrait de phase linéarisé de la Fractale de Mandelbrot, nous itérons sur différentes valeurs de $C$ dans la région d'intérêt du plan complexe et traçons les trajectoires correspondantes en utilisant une échelle de couleurs pour représenter le nombre d'itérations nécessaires pour que $|Z[n]|$ dépasse une certaine valeur limite (temps de convergence).

Voici le code Python pour générer le portrait de phase linéarisé de la Fractale de Mandelbrot :

```python
import numpy as np
import matplotlib.pyplot as plt

# Définition de la grille dans le plan complexe
xmin, xmax = -2.0, 1.0
ymin, ymax = -1.5, 1.5
resolution = 1000
x = np.linspace(xmin, xmax, resolution)
y = np.linspace(ymin, ymax, resolution)
X, Y = np.meshgrid(x, y)
C = X + 1j * Y  # Grille complexe C

# Itérations pour chaque point de la grille
max_iterations = 100
Z = np.zeros_like(C, dtype=np.complex128)
iterations = np.zeros_like(C, dtype=np.int32)

for i in range(max_iterations):
    mask = np.less(np.abs(Z), 2.0)  # Points qui n'ont pas divergé
    Z[mask] = Z[mask] + C[mask]
    iterations += mask

# Tracer le portrait de phase
plt.imshow(iterations, extent=(xmin, xmax, ymin, ymax),
           cmap='hot', origin='lower', aspect='auto')
plt.colorbar(label='Nombre d\'itérations')
plt.title('Portrait de phase linéarisé de la Fractale de Mandelbrot')
plt.xlabel('Re(C)')
plt.ylabel('Im(C)')
plt.show()
```

###A.2.5. Calcul de l’ensemble Omega-limite et Etude d’existence de cycle lime

L'ensemble $\Omega$-limite est défini comme l'ensemble des valeurs atteintes par une orbite lorsque le nombre d'itérations tend vers l'infini. Dans le cas de la Fractale de Mandelbrot, nous allons itérer sur chaque point $C$ du plan complexe et déterminer si l'orbite générée par l'équation récurrente $Z_{n+1} = Z_n^2 + C$ converge vers une valeur limite ou si elle diverge vers l'infini.

L'existence de cycles limites dans la Fractale de Mandelbrot se réfère à la présence de trajectoires périodiques ou récurrentes dans l'ensemble des orbites. Nous allons rechercher ces cycles en détectant les valeurs de $C$ pour lesquelles l'orbite finit par se répéter après un certain nombre d'itérations.

Voici un code Python pour calculer l'ensemble $\Omega$-limite et détecter les cycles limites dans la Fractale de Mandelbrot :

```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = 0
    for i in range(max_iter):
        z = z*z + c
        if abs(z) > 2:
            return i
    return max_iter

def omega_limit(c, max_iter, threshold=0.001):
    z = c
    for i in range(max_iter):
        z = z*z + c
        if abs(z - c) < threshold:
            return True
    return False

# Paramètres du calcul de l'ensemble de Mandelbrot
xmin, xmax = -2.5, 1.5
ymin, ymax = -2, 2
width, height = 1000, 1000
max_iter = 100

# Générer les points dans la grille
x = np.linspace(xmin, xmax, width)
y = np.linspace(ymin, ymax, height)
X, Y = np.meshgrid(x, y)
C = X + 1j * Y

# Calculer l'ensemble oméga-limite et les cycles limites
omega_set = set()
cycles = []
for i in range(width):
    for j in range(height):
        c = C[j, i]
        if omega_limit(c, max_iter):
            omega_set.add(c)
            if c not in cycles:
                cycles.append(c)

# Tracer l'ensemble oméga-limite
plt.scatter(np.real(list(omega_set)), np.imag(list(omega_set)), s=1, color='blue')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.title('Ensemble oméga-limite dans la fractale de Mandelbrot')
plt.show()


```

##**A.3. Analyse numérique: (module méthode numérique pour l’ingénieur)**

###A.3.1. Schéma numérique pour la résolution du système dynamique

Dans le cadre de la résolution du système dynamique de la Fractale de Mandelbrot, nous devons choisir un schéma numérique approprié pour générer les itérations nécessaires pour chaque point du plan complexe. Un schéma numérique efficace et précis est essentiel pour obtenir des résultats fiables et précis.

Pour la Fractale de Mandelbrot, un schéma numérique couramment utilisé est l'algorithme de l'itération de Mandelbrot. Cet algorithme consiste à itérer l'équation récurrente $Z_{n+1} = Z_n^2 + C$ pour chaque point $C$ du plan complexe, et à vérifier si la suite générée par cette équation diverge vers l'infini.

L'algorithme de l'itération de Mandelbrot est choisi pour plusieurs raisons :

*Simplicité :* L'algorithme est relativement simple à implémenter et à comprendre. Il ne nécessite pas de connaissances mathématiques avancées et peut être facilement traduit en code Python.

*Efficacité :* L'algorithme de l'itération de Mandelbrot est efficace en termes de temps de calcul. Il itère simplement l'équation récurrente pour chaque point du plan complexe, ce qui permet de générer rapidement les résultats.

*Précision :* L'algorithme de l'itération de Mandelbrot offre une bonne précision pour la génération de la fractale. En utilisant une résolution suffisamment élevée et un nombre d'itérations approprié, nous pouvons obtenir des détails fins et une représentation précise de la fractale.

```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = 0
    for i in range(max_iter):
        z = z*z + c
        if abs(z) > 2:
            return i
    return max_iter

def generate_mandelbrot(width, height, x_min, x_max, y_min, y_max, max_iter):
    image = np.zeros((height, width))
    for row in range(height):
        for col in range(width):
            x = np.interp(col, [0, width], [x_min, x_max])
            y = np.interp(row, [0, height], [y_min, y_max])
            c = x + y * 1j
            iteration = mandelbrot(c, max_iter)
            image[row, col] = iteration
    return image

# Paramètres de la fractale
width = 800
height = 800
x_min = -2.5
x_max = 1.5
y_min = -2.0
y_max = 2.0
max_iter = 100

# Génération de la fractale de Mandelbrot
fractal_image = generate_mandelbrot(width, height, x_min, x_max, y_min, y_max, max_iter)

# Affichage de la fractale
plt.imshow(fractal_image, cmap='hot', extent=[x_min, x_max, y_min, y_max])
plt.title('Fractale de Mandelbrot')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.show()
```

##**A.4. Implémentation :**

Dans cette section, nous allons implémenter l'algorithme de résolution pour la Fractale de Mandelbrot sous forme d'une fonction en utilisant la programmation Python. Nous identifierons également une fonction pré-définie équivalente dans Python, telle que odeint ou solve, pour résoudre le système dynamique.

Voici l'implémentation de l'algorithme de résolution de la Fractale de Mandelbrot en Python :

```python
import numpy as np

def mandelbrot_iteration(c, max_iter):
    z = 0
    for i in range(max_iter):
        z = z*z + c
        if abs(z) > 2:
            return i
    return max_iter

def mandelbrot_set(width, height, x_min, x_max, y_min, y_max, max_iter):
    image = np.zeros((height, width))
    for row in range(height):
        for col in range(width):
            x = np.interp(col, [0, width], [x_min, x_max])
            y = np.interp(row, [0, height], [y_min, y_max])
            c = x + y * 1j
            iteration = mandelbrot_iteration(c, max_iter)
            image[row, col] = iteration
    return image

# Paramètres de la fractale
width = 800
height = 800
x_min = -2.5
x_max = 1.5
y_min = -2.0
y_max = 2.0
max_iter = 100

# Résolution de la Fractale de Mandelbrot
fractal_image = mandelbrot_set(width, height, x_min, x_max, y_min, y_max, max_iter)

# Affichage de la fractale
plt.imshow(fractal_image, cmap='hot', extent=[x_min, x_max, y_min, y_max])
plt.title('Fractale de Mandelbrot')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.show()
```

Dans ce code, la fonction mandelbrot_iteration effectue l'itération de l'équation récurrente de la Fractale de Mandelbrot pour un point donné et renvoie le nombre d'itérations avant que la suite ne diverge. La fonction mandelbrot_set parcourt chaque pixel de l'image souhaitée, convertit les coordonnées du pixel en un nombre complexe c, et appelle la fonction mandelbrot_iteration pour obtenir le nombre d'itérations pour ce point. L'image finale de la fractale est stockée dans un tableau image et renvoyée à la fin.

Il convient de noter que cette implémentation est basée sur une approche directe et itérative de la Fractale de Mandelbrot. Il n'y a pas d'équation différentielle sous-jacente à résoudre, donc les fonctions odeint et solve ne sont pas applicables dans ce cas.

Pour comparer notre propre algorithme avec une méthode pré-définie telle que odeint, nous devons identifier une équation différentielle associée à la Fractale de Mandelbrot, ce qui peut être un défi car il n'y a pas d'équation différentielle directe correspondante. Par conséquent, la comparaison spécifique entre les deux méthodes n'est pas applicable dans ce contexte.

**Ensembles de Julia**


Fixons maintenant une valeur de c et ´etudions le comportement dynamique de chaque valeur de départ possible $z0 ∈ C$. On nomme la fonction de la fractale de Mandelbrot le polynôme suivant $Qc(z) = z^2 + c$

L’ensemble des valeurs de départ dont l’orbite est bornée
sous itération de $Qc$ forme l’ensemble rempli de Julia de $Qc$, que l’on note $K(Qc)$.
formellement, on a les définitions suivantes.


Soit $c ∈ C$. L’ensemble rempli de Julia de $Qc$ est:

$K(Qc) = (z : |Q^n_c
(z)$ ne diverge pas, $n → ∞)$

L’ensemble de Julia de $Qc$, noté $J(Qc)$, est la frontière de cet ensemble

Prenons $Q_0(z)=z^2$. Alors si $z_0=rexp(iθ)$, on a que $z_n=r(2^n)*exp(i2^n*θ)=Q^n_0(z_0)$

Ainsi, on a:

-$z_n → 0$  si  $r<1$

-$z_n → ∞ $ si $r>1$

-$|z_n|=1$ $∀n$ si $r=1$

Ainsi, on a partitionné $C$ en trois régions :

$E1 = ({z : |z| < 1})$

$E2 = ({z : |z| = 1})$

$E3 = ({z : |z| > 1})$


qui se distinguent par le comportement de l’it´eration de $Q_0$ sur celles-ci.

De plus, on a $K(Q0) = ({z : |z| 6 1})$ (le disque unité fermé), $J(Q0) = ({z : |z| = 1})$ (le
cercle unité)

En général, on ne peut pas trouver analytiquement l’ensemble de Julia associé à $Qc$. On doit
le déterminer numériquement à l’aide d’un ordinateur. Tout comme lorsqu’on programme
l’algorithme pour déterminer l’ensemble de Mandelbrot, on a besoin d’un critère nous permettant de déterminer à partir de quel moment l’orbite d’un point donné va tendre vers
l’infini assurément. Le corollaire suivant, nous fournit ce critère.

**Corollaire:** Soit $c ∈ C$. Supposons $|z| > max({|c|, 2})$, alors $Q^n_c (z) → ∞ $si $n → ∞$.
En particulier,$ J(Qc) ⊂ K(Qc) ⊂ ({z : |z|< max(|c|, 2)})$ .

Ce dernier résultat nous permet d’obtenir ≪l’algorithme≫ suivant pour déterminer l’ensemble
rempli de Julia $K(Qc)$ :

- On sait que $K(Qc) ⊂$ disque de rayon $max({|c|, 2})$.

- Choisissons $N$ un nombre maximum d’it´erations.

- Pour chaque point z d’une grille recouvrant le disque de rayon $max({|c|, 2})$, calculer
$Q^n_c(z)$
, $n = 1, 2, . . . , N$.


-  Si $|Q^i_c(z)| > max({|c|, 2})$  pour un $i<N$, arreter l’itération et laisser le point $z$ en
blanc

- Si $|Q^i_c(z)| < max({|c|, 2})$, $∀i< N$, colorer le point z en noir

L’ensemble des points noircis donnera une approximation de $K(Qc)$.

Voici un exemple de code Python qui utilise la bibliothèque Matplotlib pour dessiner l'ensemble de Julia pour un nombre complexe c donné :

```python
import numpy as np
import matplotlib.pyplot as plt

def julia_set(c, width, height, x_min, x_max, y_min, y_max, max_iter):
    # Créer une grille de points dans le plan complexe
    x = np.linspace(x_min, x_max, width)
    y = np.linspace(y_min, y_max, height)
    X, Y = np.meshgrid(x, y)
    Z = X + 1j * Y

    # Itérer pour chaque point de la grille
    img = np.zeros((height, width))
    for i in range(max_iter):
        Z = Z**2 + c
        mask = np.abs(Z) < 10
        img += mask

    return img

# Paramètres du dessin
c = complex(-0.7, 0.27015)
width = 800
height = 800
x_min = -1.5
x_max = 1.5
y_min = -1.5
y_max = 1.5
max_iter = 100

# Calculer l'ensemble de Julia
img = julia_set(c, width, height, x_min, x_max, y_min, y_max, max_iter)

# Afficher l'image
plt.imshow(img.T, cmap='hot', extent=(x_min, x_max, y_min, y_max))
plt.title("Ensemble de Julia pour c = {}".format(c))
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.show()
```

**Définition**: Soit $z⋆ ∈ C$ un point périodique attractif de $Qc$. Considérons son orbite $γ = ({z⋆, Qc(z⋆), . . . , Q^p_c
(z⋆)}) $.   Le bassin d’attraction de $γ$ (on dit aussi de $z⋆$) est:
$A(γ) = ({z ∈ C : d(Q^n_c
(z), γ) → 0 lorsque n → ∞})$.

Le bassin immédiat d’attraction $B(γ)$ est l’ensemble des composantes connexes de $A(γ)$ qui
contiennent $γ$.

**Théorème 1**: Soit $c ∈ C$. Soit $z⋆$ un point périodique attractif de $Qc$ et soit $γ$ son orbite.
Alors, $0$ appartient au bassin immédiat d’attraction $B(γ)$. En particulier, $d(Q^n_c
(0), γ) → 0$
lorsque $n → ∞$.

**Théorème 2 :**
L’ensemble de Julia $J(Qc)$ est connexe si et seulement si l’orbite de 0 est
bornée.

**Corollaires :**

1.    Soit $c ∈ C$. $Qc$ possède au plus un cycle attractif.
2.   Si l’orbite de 0 tend vers l’infini, alors $Qc$ n’a aucun point périodique attractif.

**Remarques** **:**
1. Ainsi, l’ensemble de Mandelbrot constitue l’ensemble des valeurs de $c ∈ C$ pour lesquelles l’ensemble de Julia $J(Qc)$ est connexe.
2. En somme, les résultats précédents montre qu’en étudiant seulement l’orbite de 0 sous $Qc$ on peut avoir une idée générale du comportement dynamique de cette fonction sur tout le plan complexe.

**Etude de composantes de l'ensemble de mandelbrot :**

Lorsque, pour un certain $c$,  $Qc$  possède un cycle attractif $γ_c$, on sait, par le théorème 1, que l’orbite de 0 se trouve dans le bassin d’attraction
B($γ_c$) et qu’elle est donc bornée. Ainsi, la valeur de $c$ en question se trouve dans l’ensemble de Mandelbrot. Dans cette section, on désire faire l’étude de certaines composantes de M de la forme:

{ $ c ∈ M$ : $Q_c$ posséde un cycle attractif de période n}.

**Cardioide principale**

Déterminons les valeurs de c pour lesquelles $Qc$ a un point fixe attractif ou neutre.

Les points fixes de $Qc$ sont les solutions de
$z = Qc(z)$ c’est-à-dire
$z = z^2 + c$ c’est-à-dire
$z^2 − z + c = 0$

D’autre part, ayant $Q′
c
(z) = 2z$, les points fixes attractifs ou neutres sont les solutions de l'équation précédente et $|2z| <1$

Ecrivons $2z$ sous forme exponentielle $2z=rexp(iθ)$

en reportant dans l'équation, on obtient:
$c=1/2rexp(iθ)-1/4r^2exp(i2θ)$

De plus, on demande
$0 < r < 1$

Autrement dit, $c ∈ C$ vérifie l'équation et la condition $|2z| <1$ si et seulement si:
$c=1/2rexp(iθ)-1/4r^2exp(i2θ)$ où $0 < r < 1$, et  $θ ∈ [0, 2π)$

Ainsi, $Qc$ a un point fixe attractif si $r < 1$ et neutre si $r = 1$.

L’ensemble des c ∈ C vérifiant  est l’ensemble borné par la courbe $c=1/2rexp(iθ)-1/4r^2exp(i2θ)$ et  $θ ∈ [0, 2π)$ en forme de cardioide.  Cet ensemble constitue le corps principal de l'ensemble de Mandelbrot M.

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 1000)  # Valeurs de l'angle theta

# Équation du cardioïde
r = (1/2) * np.exp(1j*theta) - (1/4) * np.exp(2j*theta)

x = np.real(r)  # Composante réelle de r
y = np.imag(r)  # Composante imaginaire de r

plt.plot(x, y)
plt.axis('equal')  # Pour avoir une échelle égale sur les axes x et y
plt.title('Corps de M c(θ) = (1/2)e^(iθ) - (1/4)e^(2iθ)')
plt.xlabel('Axe x')
plt.ylabel('Axe y')
plt.grid(True)
plt.show()
```

**Bulbe pour la période 2**

Les points de période 2 sont les solutions de:

$z = Q_c^2(z)$  c’est-à-dire

$z = (z^2 + c)^2 + c$ c’est-à-dire

$(z^2 + c)^2 + c − z = 0$

Les points fixes de $Q_c$ sont solutions de l'équation précèdente, mais les points fixes vérifient $z^2 − z + c = 0$.

Ainsi, on peut factoriser le polynome dans l'équation précèdente par $z^2 − z + c$. On obtient

$(z^2 + c)^2 + c − z$ $=$ $(z^2 − z + c)(z^2 + z + 1 + c)$

et ainsi les points de période 2 qui ne sont pas des points fixes sont les solutions de

$z^2 + z + 1 + c = 0$

Ayant $(Q_c^2)(z) = (z^2 + c)^2 + c$, on obtient

$(Q_c^2)'(z)= 2(z^2 + c) · 2z$

Soient $w1$ et $w2$ les solutions de $z^2 + z + 1 + c = 0$ les deux points de période $2$, c’est-à-dire

$w1 = −(1 + \sqrt{1 - 4 - 4c})/2$ $= 1/2(−1 + \sqrt{−3 −4c})$

$w2 = − (1 - \sqrt{1 - 4 - 4c})/2$ $= 1/2(−1 - \sqrt{−3 −4c})$

On a

$Q_c(w_1) = w_2$
c’est-à-dire $w_1^2 + c = w_2$

et

$Q_c(w_2) = w_1$
c’est-à-dire $w_2^2 + c = w_1$

d’où

$(Q_c^2)′(w_1) = 2(w_1^2 + c)2w_1$$= 2w_22w_1 = 4w_1w_2$

$(Q_c^2)′(w_2) = 2(w_2^2 + c)2w_2$$= 2w_12w_2 = 4w_1w_2$

Mais $w1$ et $w2$ étant les racines de $z^2 + z + 1 + c = 0$; on a

$w_1w_2 = 1 + c$

d’où

$(Q_c^2)′(w_1) = (Q_c^2)′(w_2) = 4(1+c)$

On déduit donc que $Q_c$ a un point de période 2 attractif si

$|(Q_c^2)′(w_1)| = |(Q_c^2)′(w_2)| = |4(1+c)|<1$

c’est-à-dire

$|1+c|<1/4$

donc, pour $c$ dans le disque centré en $−1$ de rayon $1/4$.

En ajoutant ce disque a la cardioide principale, on obtient les deux plus grandes composantes
de $M$.

On appellera ce disque le bulbe pour la période $2$.

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_bulbe(c):
    fig, ax = plt.subplots()
    ax.set_aspect(1)

    # Coordonnées du centre du disque
    center = complex(c, 0)

    # Rayon du disque
    radius = 1/4

    # Création du cercle
    circle = plt.Circle((center.real, center.imag), radius, edgecolor='blue', facecolor='none')

    # Ajout du cercle à la figure
    ax.add_patch(circle)

    # Configuration de la fenêtre graphique
    ax.autoscale(enable=True, axis='both', tight=True)
    ax.axhline(y=0, color='black', linewidth=0.5)
    ax.axvline(x=0, color='black', linewidth=0.5)
    ax.set_xlabel('Re')
    ax.set_ylabel('Im')
    ax.set_title(f'Bulbe pour la période 2 avec c = {c}')

    # Affichage de la figure
    plt.show()

# Appel de la fonction pour tracer le bulbe avec c = 0.5
plot_bulbe(0.5)
```
